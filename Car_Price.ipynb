```python
x = cardata.iloc[:,1].to_numpy()
y = cardata.iloc[:,2].to_numpy()

train_x,test_x,train_y,test_y= train_test_split( x, y, test_size=0.2)

train_x= train_x.reshape(-1, 1)
train_y= train_y.reshape(-1, 1)

test_x= test_x.reshape(-1, 1)
test_y= test_y.reshape(-1, 1)
# Create linear regression object
regr = linear_model.LinearRegression()
# Train the model using the training sets
regr.fit(train_x, train_y)
# Make predictions using the testing set
pred_y = regr.predict(test_x)

# The coefficients  r = sum(Zx*Zy)/n-1 
print("Coefficients: \n", regr.coef_)
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(test_y, pred_y)) #%.2f 2tnin ba3d al diecimal point
# The coefficient of determination: 1 is perfect prediction  r^2 
print("Coefficient of determination: %.2f" % r2_score(test_y, pred_y))

# Plot outputs
plt.scatter(test_x, test_y, color="gray")
plt.plot(test_x, pred_y, color="purple", linewidth=3)

plt.xticks(())
plt.yticks(())
plt.xlabel('year', fontsize = 14, fontweight ='bold')
plt.ylabel('selling_price', fontsize = 14, fontweight ='bold')
plt.title('Linear Regression\n', fontsize = 18, fontweight ='bold')
plt.show()
```

```python
x = cardata.iloc[:,3].to_numpy()
y = cardata.iloc[:,2].to_numpy()
train_x,test_x,train_y,test_y= train_test_split( x, y, test_size=0.2)

train_x= train_x.reshape(-1, 1)
train_y= train_y.reshape(-1, 1)

test_x= test_x.reshape(-1, 1)
test_y= test_y.reshape(-1, 1)
# Create linear regression object
regr = linear_model.LinearRegression()
# Train the model using the training sets
regr.fit(train_x, train_y)
# Make predictions using the testing set
pred_y = regr.predict(test_x)

# The coefficients
print("Coefficients: \n", regr.coef_)
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(test_y, pred_y))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(test_y, pred_y))

# Plot outputs
plt.scatter(test_x, test_y, color="gray")
plt.plot(test_x, pred_y, color="purple", linewidth=3)

plt.xticks(())
plt.yticks(())
plt.xlabel('km_driven', fontsize = 14, fontweight ='bold')
plt.ylabel('selling_price', fontsize = 14, fontweight ='bold')
plt.title('Linear Regression\n', fontsize = 18, fontweight ='bold')
plt.show()
```
